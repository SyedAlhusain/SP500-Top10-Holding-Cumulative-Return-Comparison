```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

#define the list of top companies
top_companies = ['AAPL', 'MSFT', 'AMZN', 'NVDA', 'GOOGL', 'TSLA', 'META', 'GOOG', 'BRK.B', 'UNH']

#fetch historical stock prices for the top companies and S&P 500
top_df = yf.download(top_companies, start='2021-01-01', end='2023-01-01')['Adj Close']
sp500_df = yf.download('^GSPC', start='2021-01-01', end='2023-01-01')['Adj Close']

#calculate daily returns for the top companies and S&P 500
top_returns = top_df.pct_change()
sp500_returns = sp500_df.pct_change()

#calculate cumulative returns
top_cumulative_returns = (1 + top_returns).cumprod() - 1
sp500_cumulative_returns = (1 + sp500_returns).cumprod() - 1

#plot the results
plt.figure(figsize=(10, 6))
for company in top_companies:
    plt.plot(top_cumulative_returns.index, top_cumulative_returns[company], label=company)

plt.plot(sp500_cumulative_returns.index, sp500_cumulative_returns, label='S&P 500', linestyle='--', color='black')

plt.title('Top 10 Holdings in the S&P 500 vs Other S&P 500 Holding - Cumulative Returns')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.show()
```

## Relative performance:

The graph tracks how the top 10 holdings have performed relative to the overall S&P 500 index over time.


## Variability

There is considerable variability in the performance of the individual holdings, with some outperforming the S&P 500 and others underperforming.


## Comparison

Nvidia (NVDA) and Google (GOOGLE) have consistently been among the top performers in comparison to S&P 500 while Amazon(AMZN) has been relative low in comparison to S&p 500 over time period 2021-2023.


## Time Trends

The graph also shows how the performance of the holdings and the index has changed over time, highlighting periods of growth and decline.


```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

#define the list of top companies
top_companies = ['AAPL', 'MSFT', 'AMZN', 'NVDA', 'GOOGL', 'TSLA', 'META', 'GOOG', 'BRK.B', 'UNH']

#fetch historical stock prices for the top companies and S&P 500
top_df = yf.download(top_companies, start='2021-01-01', end='2023-01-01')['Adj Close']
sp500_df = yf.download('^GSPC', start='2021-01-01', end='2023-01-01')['Adj Close']

#calculate daily returns for the top companies and S&P 500
top_returns = top_df.pct_change()
sp500_returns = sp500_df.pct_change()

#calculate average return for the top companies
average_top_returns = top_returns.mean(axis=1)

#calculate cumulative returns
top_cumulative_returns = (1 + average_top_returns).cumprod() - 1
sp500_cumulative_returns = (1 + sp500_returns).cumprod() - 1

#plot the results
plt.figure(figsize=(10, 6))
plt.plot(top_cumulative_returns.index, top_cumulative_returns, label='Top 10 Companies (Average)', linestyle='--', color='blue')
plt.plot(sp500_cumulative_returns.index, sp500_cumulative_returns, label='S&P 500', linestyle='--', color='black')

plt.title('Top 10 Holdings in the S&P 500 vs Other S&P 500 Holding - Cumulative Returns')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.show()
```

## Summary

Overall, the top 10 holdings in the S&P 500 have generally outperformed the broader index and the range of other holdings over the past two years. Their cumulative return is consistently above S&P 500 range.  It suggests that these select holdings have not only weathered market fluctuations effectively but have also contributed significantly to the positive trajectory of the S&P 500.
